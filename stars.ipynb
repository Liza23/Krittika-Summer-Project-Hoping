```
data_class0 = data[data['mag']>18.0]
data_class2 = data[(data['mag'] < 18.0) & (data['mag'] > 15.0)] 
data_class2 = data[(data['mag'] < 15.0) & (data['mag'] > 12.0)] 
data_class3 = data[(data['mag'] < 12.0) & (data['mag'] > 10.0)] 
data_class4 = data[(data['mag'] < 10.0) & (data['mag'] > 8.0)] 
data_class5 = data[(data['mag'] < 8.0) & (data['mag'] > 5.0)] 
data_class6 = data[(data['mag'] < 5.0) & (data['mag'] > 0.0)] 
```

```
plt.xlim(0,360)
plt.ylim(-90,90)
plt.xlabel('Right ascension')
plt.ylabel('Decilination')
plt.scatter(data_class0['ra'], data_class0['dec'], s = 15, color = 'blue')
plt.scatter(data_class1['ra'], data_class1['dec'], s = 10, color = 'blue')
plt.scatter(data_class2['ra'], data_class2['dec'], s = 1, color = 'blue')
plt.scatter(data_class3['ra'], data_class3['dec'], s = 0.01, color = 'blue')
plt.scatter(data_class4['ra'], data_class4['dec'], s = 0.001, color = 'blue')
plt.scatter(data_class5['ra'], data_class5['dec'], s = 0.00001, color = 'blue')
plt.scatter(data_class6['ra'], data_class6['dec'], s = 0.000001, color = 'blue')
plt.gcf().set_size_inches((10, 10))    
plt.show
```
